What is the state that spends the most on healthcare and social assistance?

```python
hdf = df[df['NAICS Description'] == 'Information']
sdf = hdf.groupby('State').sum()
sdf
```

```python
state = hdf[hdf['Annual Payroll'] == hdf['Annual Payroll'].max()]
state
```

What is the average number of people employed in information per state?

```python
idf = df[df['NAICS Description'] == 'Information']
idf
```

```python
iidf=idf.groupby('State').sum()
iidf
```

```python
average = iidf['Employment'].mean()
average
```
